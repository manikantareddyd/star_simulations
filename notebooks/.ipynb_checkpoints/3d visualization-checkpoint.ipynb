Lets create a 3d scatter plot 

```python
%pylab inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(snapshot['vx'], snapshot['vy'], snapshot['vz'])
```
